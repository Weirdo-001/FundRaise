```python
from ultralytics import YOLO

# This uses the official pretrained YOLOv8n model safely
model = YOLO('yolov8n.yaml')  # no need for .pt file
```

```python
data_yaml_content = """
train: /content/datasets/datasets/train/images
val: /content/datasets/datasets/val/images
test: /content/datasets/datasets/test/images
nc: 7
names: ['OxygenTank', 'NitrogenTank', 'FirstAidBox', 'FireAlarm', 'SafetySwitchPanel', 'EmergencyPhone', 'FireExtinguisher']
"""

data_yaml_path = "/content/datasets/data.yaml"
with open(data_yaml_path, "w") as f:
    f.write(data_yaml_content)

print("✅ data.yaml created at", data_yaml_path)
```

```python
# Train YOLOv8 on your dataset


# Train the model
model.train(
    data='/content/datasets/data.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    name="Hackathon_YOLOv8"
)

```
